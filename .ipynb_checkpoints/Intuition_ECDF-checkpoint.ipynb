<h2>Intuition behind ECDF (<code><i>Empirical distribution function</i></code>)</h2>
<p>© Vytautas Bielinskas</p>

<p>A cummulative distribution function (CDF) give the probability than a random variable X is less that a given value x. (F(x) = Pr{X<=x}).</p>

<h2># Performing ECDF</h2>

<center><i>Function: ECDF</i></center>

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""

    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, len(x)+1) / n

    return x, y
```

<center><i>Function: Calculate quartiles (percentiles)</i></center>

```python
def quartiles(data, threshold):
    # Calculate quartiles of the values
    quarts = df_['PSQMY'].quantile([threshold, 1 - threshold])
    quart_1 = quarts.iloc[0]
    quart_2 = quarts.iloc[1]
    
    return quart_1, quart_2
```

<center><i>- - - Analyzing - - -</i></center>

<b># Input the <code>threshold</code> value for data cleaning.</b>

<code>threshold</code> must be less than <b>0.5</b>. <code>0.1</code> correspond to <b>10 %</b>, <code>0.25</code> correspond to <b>25 %</b> and so on.

```python
try:
    threshold = float(input('\nInput the value for threshold.\n'))
    print('You entered {} ({}) for threshold value.'.format(threshold,
                                                           type(threshold)))
except:
    print('\nFormat of value is not float or a number.')
```

<b># Calculate percentiles of the raw dataset</b>

```python
quart_1, quart_2 = quartiles(df_, threshold)
print('Percentiles for the data in given threshold of {} is:\n|{} % : {}\n|{} % : {}'.format(threshold,
                                                                                         threshold * 100,
                                                                                         quart_1,
                                                                                         (1 - threshold) * 100,
                                                                                         quart_2))
```

<b># Plot ECDF for raw dataset</b>

```python
def plot_ecdf_distribution(data, quart_1, quart_2, avg):

    import matplotlib.patches as mpatches

    # Set color gamma manually
    primary_color   = '#1F00AA'
    secondary_color = '#F74600'
    value_color     = '#2F1557'
    subcolor        = '#32B132'

    x, y = ecdf(data['PSQMY'])

    # Generate plot
    sns.set_style('whitegrid')

    # Set plot size
    _ = plt.rcParams['figure.figsize'] = (8,5)

    _ = plt.plot(x, y, marker='.',
                 linestyle='none',
                 color = value_color,
                 alpha = 0.8)

    # Make the margins nice
    _ = plt.margins(0.02)
    
    # Overlay percentiles as red diamonds.
    percentiles = np.array([50])
    df_perc = np.percentile(data['PSQMY'],
                            percentiles)

    # Label the axes
    _ = plt.xlabel('PSQMY in Vilnius (Lithuania)\nMedium is {}\nAverage is {}.\nDifference is {}%'.format(df_perc,
                                                                                                             avg,
                                                                                                             (avg - df_perc)/avg*100),
                   fontsize = 12,
                   family='Arial')
    
    _ = plt.ylabel('ECDF',
                   fontsize = 12)

    _ = plt.plot(df_perc, percentiles/100, marker='D', color='red', linestyle='none', markersize = 10)

    # Draw vertical line on a plot
    _ = plt.axvline(x = df_perc, color = primary_color)
    _ = plt.axvline(x = avg, color = subcolor)
    _ = plt.axvline(x = quart_1, color = secondary_color, linestyle = '--')
    _ = plt.axvline(x = quart_2, color = secondary_color, linestyle = '--')

    # Draw a horizontal line on a plot
    _ = plt.axhline(y = 0.5, color = primary_color)
    _ = plt.axhline(y = threshold, color = secondary_color, linestyle = '--')
    _ = plt.axhline(y = 1 - threshold, color = secondary_color, linestyle = '--')

    plt.tick_params(labelsize=12)

    # Set axis style
    _ = plt.grid(b=True,
                 which='major',
                 color='#cccccc',
                 linestyle='--')

    # Add legend
    patch = mpatches.Patch(color = value_color,
                           label='Dataset')

    _ = plt.legend(handles=[patch],
                   loc = 'lower right',
                   fontsize = 14)

    # Display the plot
    plt.show()
    
    return None
```

```python
plot_ecdf_distribution(df_, quart_1, quart_2, avg_1)
```

<b># Removing outliers (over and below <code>Threshold</code>)</b>

```python
len_1 = len(df_)

df_ = df_[df_['PSQMY'] > quart_1]
df_ = df_[df_['PSQMY'] < quart_2]

len_2 = len(df_)
print('After data cleaning we lost {} values and now having {} ones.'.format(len_1 - len_2,
                                                                            len_2))
```

<b># Calculate percentiles of cleaned dataset</b>

```python
quart_1, quart_2 = quartiles(df_, threshold)
print('Percentiles for the data in given threshold of {} is:\n|{} % : {}\n|{} % : {}'.format(threshold,
                                                                                         threshold * 100,
                                                                                         quart_1,
                                                                                         (1 - threshold) * 100,
                                                                                         quart_2))
```

```python
plot_ecdf_distribution(df_, quart_1, quart_2, df_['PSQMY'].mean())
```
